```python
import pandas as pd 
dfs=pd.read_csv("commoncore.0.train.csv")
dfs2=pd.read_csv("commoncore.0.valid.csv")
dfs3=pd.read_csv("commoncore.0.test.csv")
```

```python
totalDf=pd.DataFrame()
for i in range(0,5):
    print(len(dfs[dfs['Grade']==i]))
    print(len(dfs2[dfs2['Grade']==i]))
    print(len(dfs3[dfs3['Grade']==i]))
    totalDf=totalDf.append(dfs[dfs['Grade']==i])
    totalDf=totalDf.append(dfs2[dfs2['Grade']==i])
    totalDf=totalDf.append(dfs3[dfs3['Grade']==i])
```

```python
average=0
for i in totalDf[totalDf['Grade']==0]['Text']:
    print(i.split())
    print(len(i.split()))
    average=average+len(i.split())
print(average)
print(len(totalDf[totalDf['Grade']==0]['Text']))
print(average/len(totalDf[totalDf['Grade']==0]['Text']))
```
