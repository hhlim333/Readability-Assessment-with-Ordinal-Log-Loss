```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.metrics import max_error
acc_list=[]
adjacent_accuracy_list=[]
f1_list=[]
pre_list=[]
rec_list=[]
qwk_list=[]
for i in range(0,5):
    dfs=pd.read_csv("mainland.chinese-xlnet-base."+str(i)+".14.csv")
    #dfs['FinalLabel'].value_counts()[True]/len(dfs['FinalLabel'])
    adjacent_accuracy = np.sum(np.abs(dfs['prediction'] - dfs['label']) <= 1) / len(dfs['prediction'] )
    acc = sklearn.metrics.accuracy_score(dfs['label'], dfs['prediction'])
    f1 = sklearn.metrics.f1_score(dfs['label'], dfs['prediction'], average='weighted')
    pre = sklearn.metrics.precision_score(dfs['label'], dfs['prediction'] ,average='weighted')
    rec = sklearn.metrics.recall_score(dfs['label'],dfs['prediction'], average='weighted')
    qwk = sklearn.metrics.cohen_kappa_score(dfs['label'],dfs['prediction'], weights="quadratic")
    acc_list.append(acc)
    adjacent_accuracy_list.append(adjacent_accuracy)
    f1_list.append(f1)
    pre_list.append(pre)
    rec_list.append(rec)
    qwk_list.append(qwk)
    print(f"Test Adjacent Accuracy: {adjacent_accuracy * 100:.2f}%")
    print(f"Test Accuracy: {acc * 100:.2f}%")
    print(f"Test F1-score: {f1 * 100:.2f}%")
    print(f"Test Precision: {pre * 100:.2f}%")
    print(f"Test Recall: {rec * 100:.2f}%")
    print(f"Test QWK: {qwk * 100:.2f}%")
print("avg adjacent_accuracy:" +str(round(float(sum(adjacent_accuracy_list) / len(adjacent_accuracy_list)), 3)))
print("avg acc:" +str(round(float(sum(acc_list) / len(acc_list)), 3)))
print("avg f1:" +str(round(float(sum(f1_list) / len(f1_list)), 3)))
print("avg pre:" +str(round(float(sum(pre_list) / len(pre_list)), 3)))
print("avg rec:" +str(round(float(sum(rec_list) / len(rec_list)), 3)))
print("avg qwk:" +str(round(float(sum(qwk_list) / len(qwk_list)), 3)))
```

```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.metrics import max_error
acc_list=[]
adjacent_accuracy_list=[]
f1_list=[]
pre_list=[]
rec_list=[]
qwk_list=[]
for i in range(0,5):
    dfs=pd.read_csv("mainland."+str(i)+".test.csv")
    dfs['prediction']=dfs['chinese-xlnet-base.14.pred']
    dfs['label']=dfs['Grade']
    #dfs['FinalLabel'].value_counts()[True]/len(dfs['FinalLabel'])
    adjacent_accuracy = np.sum(np.abs(dfs['prediction'] - dfs['label']) <= 1) / len(dfs['prediction'] )
    acc = sklearn.metrics.accuracy_score(dfs['label'], dfs['prediction'])
    f1 = sklearn.metrics.f1_score(dfs['label'], dfs['prediction'], average='weighted')
    pre = sklearn.metrics.precision_score(dfs['label'], dfs['prediction'] ,average='weighted')
    rec = sklearn.metrics.recall_score(dfs['label'],dfs['prediction'], average='weighted')
    qwk = sklearn.metrics.cohen_kappa_score(dfs['label'],dfs['prediction'], weights="quadratic")
    acc_list.append(acc)
    adjacent_accuracy_list.append(adjacent_accuracy)
    f1_list.append(f1)
    pre_list.append(pre)
    rec_list.append(rec)
    qwk_list.append(qwk)
    print(f"Test Adjacent Accuracy: {adjacent_accuracy * 100:.2f}%")
    print(f"Test Accuracy: {acc * 100:.2f}%")
    print(f"Test F1-score: {f1 * 100:.2f}%")
    print(f"Test Precision: {pre * 100:.2f}%")
    print(f"Test Recall: {rec * 100:.2f}%")
    print(f"Test QWK: {qwk * 100:.2f}%")
print("avg adjacent_accuracy:" +str(round(float(sum(adjacent_accuracy_list) / len(adjacent_accuracy_list)), 3)))
print("avg acc:" +str(round(float(sum(acc_list) / len(acc_list)), 3)))
print("avg f1:" +str(round(float(sum(f1_list) / len(f1_list)), 3)))
print("avg pre:" +str(round(float(sum(pre_list) / len(pre_list)), 3)))
print("avg rec:" +str(round(float(sum(rec_list) / len(rec_list)), 3)))
print("avg qwk:" +str(round(float(sum(qwk_list) / len(qwk_list)), 3)))
```
